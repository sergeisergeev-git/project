```python
from datetime import datetime

def validate_date(date_string):
    """
    Валидирует строку даты. Проверяет формат ДД.ММ.ГГГГ и гарантирует,
    что дата не меньше текущей даты.
    """
    try:
        # Преобразование строки в объект даты
        parsed_date = datetime.strptime(date_string, '%d.%m.%Y').date()
    except ValueError:
        return False

    # Текущая системная дата
    current_date = datetime.today().date()

    # Проверка: дата должна быть сегодня или позже
    return parsed_date >= current_date

# Примеры использования с учетом текущей даты (2026 год):
print(f"Сегодняшняя дата: {datetime.today().date().strftime('%d.%m.%Y')}")
print(f"Валидация '25.12.2027' (Будущее): {validate_date('25.12.2027')}")
print(f"Валидация '01.01.2020' (Прошлое): {validate_date('01.01.2020')}")
print(f"Валидация 'некорректный формат': {validate_date('31-12-2027')}")
```
